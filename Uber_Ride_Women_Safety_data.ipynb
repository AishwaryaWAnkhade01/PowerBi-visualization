## Visualization 

### 1. Time of Day and Safety Risk Correlation


#### Objective: Determine if there are specific times of day when women face higher risks.

#### Insight:
##### By comparing incident reports across different time intervals (morning, afternoon, evening, night), we can identify riskier times. If incidents are more common late at night, this indicates the need for more security measures during these hours.


#### Visualization: Line plot or heatmap showing incident rates across different hours of the day, segmented by gender.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'START_DATE' is in datetime format
cleaned_df['START_DATE'] = pd.to_datetime(cleaned_df['START_DATE'])

# Convert 'Yes'/'No' in 'INCIDENT_REPORT' to 1/0
cleaned_df['INCIDENT_REPORT'] = cleaned_df['INCIDENT_REPORT'].map({'Yes': 1, 'No': 0})

# Extract hour of the day from 'START_DATE'
cleaned_df['TIME_OF_DAY'] = cleaned_df['START_DATE'].dt.hour

# Group by 'hour_of_day' and 'RIDER_GENDER' to calculate total incidents
incident_by_hour_gender = cleaned_df.groupby(['TIME_OF_DAY', 'RIDER_GENDER'])['INCIDENT_REPORT'].sum().unstack(fill_value=0)

# --- Line Plot ---
plt.figure(figsize=(12, 6))
sns.lineplot(data=incident_by_hour_gender, markers=True, dashes=False)
plt.title('Incident Rates Across Different Hours of the Day by Gender')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Incidents')
plt.legend(title='Gender')
plt.show()

# --- Heatmap ---
plt.figure(figsize=(12, 6))
sns.heatmap(incident_by_hour_gender, annot=True, cmap='YlGnBu', fmt='.0f', cbar_kws={'label': 'Number of Incidents'})
plt.title('Heatmap of Incident Rates by Hour and Gender')
plt.xlabel('Gender')
plt.ylabel('Hour of the Day')
plt.show()

# Confirmation message
print("Visualizations of incident rates by hour and gender generated successfully.")
```

### Result/Interpretation:

##### The heatmap (or line plot) indicates that incidents are more frequent during late-night hours (e.g., 10 PM - 2 AM), suggesting that late-night rides are associated with higher safety risks for women. In contrast, rides during midday (12 PM - 2 PM) show lower risk, highlighting that the timing of the ride is a crucial factor. These findings imply that additional safety measures, such as real-time monitoring or safety alerts during the late-night hours, could be beneficial for women riders.

### 6. Distance of Trip and Incidents Reported

#### Objective: Analyze if trip duration or distance correlates with higher incident frequency.



### Insight: 
##### If longer rides tend to have more incidents, this could suggest that Uber should encourage shorter trips or provide alerts for longer rides.


```python
# Convert 'Yes'/'No' values in 'INCIDENT_REPORT' to 1/0
cleaned_df['INCIDENT_REPORT'] = cleaned_df['INCIDENT_REPORT'].map({'Yes': 1, 'No': 0})

# Ensure 'MILES' is numeric
cleaned_df['MILES'] = pd.to_numeric(cleaned_df['MILES'], errors='coerce')

# Drop rows with NaN values in the 'MILES' or 'INCIDENT_REPORT' columns for cleaner analysis
cleaned_df = cleaned_df.dropna(subset=['MILES', 'INCIDENT_REPORT'])

# Check the first few rows of the cleaned data
print(cleaned_df.head())
# Define distance ranges (e.g., 0-5 miles, 5-10 miles, etc.)
distance_bins = [0, 5, 10, 15, 20, 25, 30, 35, 40]
distance_labels = ['0-5 miles', '5-10 miles', '10-15 miles', '15-20 miles', '20-25 miles', '25-30 miles', '30-35 miles', '35-40 miles']

cleaned_df['DISTANCE_RANGE'] = pd.cut(cleaned_df['MILES'], bins=distance_bins, labels=distance_labels)

# Group by 'DISTANCE_RANGE' and 'INCIDENT_REPORT' to count incidents
incident_by_distance = cleaned_df.groupby(['DISTANCE_RANGE', 'INCIDENT_REPORT']).size().unstack(fill_value=0)

# Plot the data using a bar plot
incident_by_distance.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set3')

plt.title('Figure: Incidents by Trip Distance Ranges')
plt.xlabel('Trip Distance Range')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.legend(title='Incident Report', labels=['No Incident', 'Incident'])
plt.show()

# Confirmation message
print("Bar chart showing incidents by trip distance ranges generated successfully.")
```

#### Result/Interpretation 

##### Positive Correlation with Distance: If longer trips are shown to have higher incident frequencies, it indicates that riders may face more risks as trips increase in distance.
##### Distance Threshold: The analysis may reveal a specific distance range where incidents significantly increase. For instance, rides above 10 miles might show a spike in incident reports compared to shorter rides.

#### Conclusion :
##### Longer Rides Have More Incidents: If a positive trend is observed, Uber could consider implementing additional safety features such as ride alerts for longer trips or emergency response options for extended rides.
##### Shorter Rides Are Safer: If shorter rides have relatively fewer incidents, riders might be encouraged to choose shorter trips when possible.

 ### 8.Heatmap of Risk Score by Pickup Area and Time of Day

#### Objective:
##### To analyze the risk levels associated with different pickup areas during various times of the day, based on a combination of safety scores and incident reports. This helps identify high-risk areas and times where women may face increased safety concerns.

#### Insight:
##### High-Risk Zones: The heatmap highlights pickup areas with elevated risk scores, particularly at specific times of the day. These areas may require additional safety measures, such as increased surveillance or safer ride options.
##### Time-Sensitive Risks: Certain times, such as late at night or early morning, may exhibit higher risks regardless of the pickup area, indicating time-specific vulnerabilities for riders.
##### Risk Hotspots: By combining incident reports and area safety scores, this visualization provides a detailed view of "hotspots" that may require immediate attention.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a 'Risk_Score' column by normalizing and combining relevant factors
cleaned_df['Normalized_Safety_Score'] = (1 - df['AREA_SAFETY_SCORE'] / cleaned_df['AREA_SAFETY_SCORE'].max())  # Higher score = higher risk
cleaned_df['Normalized_Incident'] =cleaned_df ['INCIDENT_REPORT'] / cleaned_df['INCIDENT_REPORT'].max()
cleaned_df['Risk_Score'] = (cleaned_df['Normalized_Safety_Score'] + cleaned_df['Normalized_Incident']) / 2  # Average of normalized values

# Group by time of day and pickup area to get the average risk score for each combination
risk_heatmap_data = cleaned_df.pivot_table(index='PICKUP_AREA', columns='TIME_OF_DAY', values='Risk_Score', aggfunc=np.mean)

# Plot the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(risk_heatmap_data, annot=True, cmap='Reds', fmt=".2f", cbar_kws={'label': 'Risk Score'})
plt.title('Figure 21: Heatmap of Risk Score by Pickup Area and Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Pickup Area')
plt.show()

# Print confirmation message
print("Risk score heatmap by pickup area and time of day generated successfully.")
```

##### Areas with the darkest shades of red represent the highest risk scores, suggesting that riders may face greater safety concerns during these times.
